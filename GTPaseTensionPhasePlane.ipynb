```python
import matplotlib.pyplot as plt
import matplotlib.cm as cm
import numpy as np
import seaborn as sns
import pandas as pd
from tqdm.notebook import tqdm, trange

rng = np.random.default_rng()

sns.set_style('ticks',{'axes.edgecolor': '[0,0,0]','xtick.direction':'in','ytick.direction':'in','ytick.right':'True','xtick.top':'True'})

plt.rcParams['text.usetex'] = False
plt.rcParams['pdf.fonttype'] = 42
plt.rcParams['ps.fonttype'] = 42
plt.rcParams['mathtext.fontset'] = 'cm'
plt.rcParams['figure.autolayout'] = 'True'
%config Completer.use_jedi = False
```

```python
beta=0.171
b=0.17

gamma=1.5
G_T=2
ell0=1
phi=0.75
G_h=0.3
epsilon=0.1
alpha=10
n=4


sns.set_context('talk')
num = 200
gmin = 0.2
gmax = 1.2
lmin = 0.15
lmax = 0.4
G,L = np.meshgrid(np.linspace(gmin,gmax,num),np.linspace(lmin,lmax,num))

gdot = (b + beta/(1+np.exp(-alpha*(L-(ell0 - phi*G**n/(G_h**n+G**n))))) + gamma*G**n/(1+G**n))*(G_T-G) - G
ldot = -epsilon*(L-(ell0 - phi*G**n/(G_h**n+G**n)))

fig,ax = plt.subplots(1,1,figsize=(6,6))
ax.streamplot(G,L,gdot,ldot,density=1)
ax.contour(G,L,gdot,levels=[0])
ax.contour(G,L,ldot,levels=[0])
ax.grid()
ax.set_xlim(gmin,gmax)
ax.set_ylim(lmin,lmax)
ax.set_xlabel('G')
ax.set_ylabel('L')
```
